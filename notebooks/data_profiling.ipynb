## 🔍 Date Field Inspection

Examine date fields and their formats prior to transformation.

```python
# Display sample values from date-related columns

df[["ScheduledDay", "AppointmentDay"]].head()
```

```python
# Convert date columns to datetime format

df["ScheduledDay"] = pd.to_datetime(df["ScheduledDay"])
df["AppointmentDay"] = pd.to_datetime(df["AppointmentDay"])

# Verify updated data types

df[["ScheduledDay", "AppointmentDay"]].dtypes
```

## 🔍 Waiting Time Analysis

Calculate appointment waiting times and assess their distribution.

```python
# Calculate waiting time in days

df["WaitingDays"] = (
    df["AppointmentDay"] - df["ScheduledDay"]
).dt.days

# Display summary statistics

df["WaitingDays"].describe()
```

```python
# Count records with negative waiting times

negative_waiting_count = (df["WaitingDays"] < 0).sum()

print(f"Negative Waiting Time Records: {negative_waiting_count:,}")
```

```python
# Display sample records with negative waiting times

df.loc[df["WaitingDays"] < 0,
       ["ScheduledDay", "AppointmentDay", "WaitingDays"]].head(10)
```

```python
# Calculate waiting days using date values only

df["WaitingDays_Corrected"] = (
    df["AppointmentDay"].dt.normalize() -
    df["ScheduledDay"].dt.normalize()
).dt.days

# Display summary statistics

df["WaitingDays_Corrected"].describe()
```

```python
# Count records with genuinely negative waiting times

negative_waiting_corrected = (
    df["WaitingDays_Corrected"] < 0
).sum()

print(
    f"Negative Waiting Time Records (Corrected): "
    f"{negative_waiting_corrected:,}"
)
```

```python
# Display records with genuine negative waiting times

df.loc[
    df["WaitingDays_Corrected"] < 0,
    [
        "ScheduledDay",
        "AppointmentDay",
        "WaitingDays_Corrected",
        "PatientId",
        "AppointmentID"
    ]
]
```

## 🔍 No-Show Status Analysis

Review the target variable and verify appointment attendance categories.

```python
df["No-show"].unique()
```

```python
# Calculate appointment attendance distribution

df["No-show"].value_counts()
```

```python
# Calculate appointment attendance percentages

df["No-show"].value_counts(normalize=True) * 100
```

## 🔍 Appointment Attendance KPIs

Calculate key appointment attendance metrics for business reporting and dashboarding.

```python
# Calculate appointment attendance KPIs

total_appointments = len(df)

attendance_rate = (
    (df["No-show"] == "No").mean() * 100
)

no_show_rate = (
    (df["No-show"] == "Yes").mean() * 100
)

print(f"Total Appointments: {total_appointments:,}")
print(f"Attendance Rate: {attendance_rate:.2f}%")
print(f"No-Show Rate: {no_show_rate:.2f}%")
```
